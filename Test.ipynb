```python
import numpy as np
from PIL import Image
import os

os.makedirs('./resources', exist_ok=True)

# Read and generate an image to verify
images = np.load('./TOFDSRD/test_images_split.npy')
print(images.shape)
img_uint8 = (images[45]).clip(0, 255).astype(np.uint8)

img_uint8 = np.transpose(img_uint8, (1,2,0))

img = Image.fromarray(img_uint8)
img.save('./resources/sample_rgb.png')
```
